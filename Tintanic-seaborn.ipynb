# Visualizaing Titanic Datas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv("train.csv")

df.head()
```

```python
sns.barplot(x = 'Sex' ,y = 'Fare', data = df)
```

```python
sns.barplot(x = 'Pclass', y = 'Fare' , hue='Sex', data=df , palette = 'autumn')
#by default estimator is mean
```

```python
sns.barplot(x = 'Pclass' , y = 'Age' ,hue = 'Sex' , data = df , palette = 'winter_r')
#by default estimator is mean
```

```python
sns.barplot(x = 'Pclass' , y = 'Age' ,hue = 'Sex' , data = df , palette = 'autumn' ,estimator = np.median)
```

```python
#filling nan with median 
df.fillna(np.nanmedian(df['Age']), inplace = True)
sns.distplot(df['Age'])
```

```python
sns.barplot(x = 'Sex', y = 'Age' , hue = 'Embarked', data = df)
```

```python
sns.boxplot( x = df['Pclass'] , y = df['Fare']  , hue = df['Sex'])
```

```python
sns.stripplot(x = df['Sex'] , y = df['Fare'] ,jitter = 0.5)
```

```python
df.tail()
```

```python
sns.barplot( x = 'Sex' , y = 'Survived' , data = df)
```

```python
sns.barplot( x = 'Sex' , y = 'Survived' , hue = 'Pclass', data = df)
```

```python
x = sns.FacetGrid(df,row = 'Sex',col = 'Pclass')
x = x.map(sns.barplot , 'Fare')
```

```python
x = sns.FacetGrid(df , row = 'Embarked' , col = 'Sex')
x = x.map(sns.barplot , 'Survived')
```

```python
y = sns.FacetGrid(df, col = 'Survived')
y.map(plt.hist, 'Fare' , bins=10)
```

```python
sns.barplot(x = 'Sex', y = 'Age' , hue = 'Survived' , data = df)
```

```python
x = sns.FacetGrid(df ,row = 'Sex' , col = 'Survived' , hue = 'Embarked' , aspect = 2.2) 
x = x.map(sns.barplot , 'Age' )
x.add_legend()
```
